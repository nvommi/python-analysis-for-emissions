# Import Libraries


```python
#data analysis
import pandas as pd
#data visualization
import matplotlib.pyplot as plt
import plotly.express as px

#download files to our computer
from google.colab import files
```

#Import Data
download data from this github [repository](https://github.com/nvommi/python-analysis-for-emissions)


```python
#import csv of emissions data

#save github csv link
emissions_data = "https://raw.githubusercontent.com/nvommi/python-analysis-for-emissions/main/final%20project%20excel%20sheet.csv"
#define our initial data frame
df_emissions = (pd.read_csv(emissions_data))
df_emissions.drop([0])
```

#Data Visualizations with Plotly

```python
#make line graph in plotly express
personal_vehicles_vs_co2 = px.scatter(data_frame=df_emissions, x = "Miles traveled", y = "Metric tons of carbon (millions)", title = "Miles Traveled vs Co2", trendline="ols")
personal_vehicles_vs_co2
```

```python
personal_vehicles_vs_co2 = px.scatter(data_frame=df_emissions, x = "vehicles", y = "Metric tons of carbon (millions)", title = "Personal Vehicles vs Co2", trendline="ols")
personal_vehicles_vs_co2
```

```python
personal_vehicles_vs_co2 = px.scatter(data_frame=df_emissions, x = "transportation", y = "Metric tons of carbon (millions)", title = "Public Transportation vs Co2", trendline="ols")
personal_vehicles_vs_co2
```
